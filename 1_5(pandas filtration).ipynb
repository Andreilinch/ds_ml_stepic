```python
# У какой доли студентов из датасэта в колонке lunch указано free/reduced?
student_Performance[student_Performance.lunch == 'free/reduced'].count()/1000
```

```python
# Как различается среднее и дисперсия оценок по предметам у групп студентов со стандартным или урезанным ланчем?
student_Performance.groupby('lunch').describe()
```

```python
# уберу пробелы
student_Performance.columns = [x.replace(" ", "_") for x in student_Performance.columns]
student_Performance
```

```python
# можно все фильтрации сделать через специализированную конструкцию .query()
student_Performance.query("writing_score > 74 & math_score > math_score.mean()")
```

```python
current_writing_score = 78
student_Performance.query("writing_score > @current_writing_score")
```

```python
# задание
student_Performance[student_Performance['parental_level_of_education'].isin(["bachelor's degree", "master's degree"])]
# отберёт студентов, уровень образования родителей которых бакалавр или магистр
```

```python
# наблюдения, оценка за математику которых выше 90 и за чтение меньше либо равна 72
student_Performance[(student_Performance['math_score'] > 90) & (student_Performance['reading_score'] <=72)]
```
