```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
```

```python
student_Performance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
student_Performance.columns = [x.replace(" ", "_") for x in student_Performance.columns]
student_Performance
```

```python
student_Performance.groupby('gender').mean()
```

```python
student_Performance.groupby('gender').aggregate({'math_score': 'mean', 'reading_score': 'mean'})
```

```python
student_Performance.groupby('gender', as_index=False) \
    .aggregate({'math_score': 'mean', 'reading_score': 'mean'}) \
    .rename(columns = {'math_score' : 'mean_math_score', 'reading_score': 'mean_reading_score'})
```

```python
# если несколько переменных - оборачиваем их в список
student_Performance.groupby(['gender','race/ethnicity'] , as_index=False) \
    .aggregate({'math_score': 'mean', 'reading_score': 'mean'}) \
    .rename(columns = {'math_score' : 'mean_math_score', 'reading_score': 'mean_reading_score'})
```

```python
# multi index, например для вставки в каонй нить отчет 
student_Performance.groupby(['gender','race/ethnicity']) \
    .aggregate({'math_score': 'mean', 'reading_score': 'mean'}) \
    .rename(columns = {'math_score' : 'mean_math_score', 'reading_score': 'mean_reading_score'})
```

```python
# сколько уникальных значений
# multi index удобно сделать для это задачи так как это сгруппированная серия
student_Performance.groupby(['gender','race/ethnicity']).math_score.nunique()
```

```python
# топ 5 студентов по gender
student_Performance.sort_values(['gender', 'math_score'], ascending=False)\
    .groupby('gender').head(5)
```

```python
# создание новых колонок
student_Performance['total_score'] = student_Performance.math_score+student_Performance.reading_score+student_Performance.writing_score
student_Performance
```

```python
# для более сложных преобразований используем метод .assign
student_Performance = student_Performance.assign(total_score_log = np.log(student_Performance.total_score))
student_Performance
```

```python
student_Performance.drop(['total_score'],axis = 1)
```

```python
concentrations = pd.read_csv('http://stepik.org/media/attachments/course/4852/algae.csv')
concentrations
# Помогите найти среднюю концентрацию каждого из веществ в каждом из родов (колонка genus)
```

```python
concentrations.groupby('genus').aggregate({'sucrose':'mean','alanin':'mean', 'citrate':'mean', 'glucose':'mean', 'oleic_acid':'mean'})
```

```python
# Пользуясь предыдущими данными, укажите через пробел (без запятых) чему равны минимальная, 
# средняя и максимальная концентрации аланина (alanin) среди видов рода Fucus. Округлите до 2-ого знака, десятичным разделителем является точка.

concentrations.query("genus == 'Fucus'").describe()
concentrations.query('genus=="Fucus"')[['alanin']].describe().round(2)
concentrations.groupby('genus').agg(['min', 'mean', 'max']).loc['Fucus', 'alanin'].round(2)
concentrations.groupby('genus').aggregate({'alanin': ['min', 'mean', 'max']}).loc[['Fucus']]

```

```python
concentrations.groupby('group').describe()
```

```python
# Число видов бурых, зелёных и красных водорослей
# Размах (максимум - минимум) сахарозы в бурых, зелёных и красных водорослях
# Дисперсия концентрации цитрата в бурых, зелёных и красных водорослях

concentrations.groupby('group', as_index = False)\
    .agg({'glucose':'count',
          'sucrose': lambda x: x.max() - x.min(),
          'citrate':'var'}) \
    .rename(columns = {'glucose':'count','sucrose':'range','citrate':'variance'})
```
